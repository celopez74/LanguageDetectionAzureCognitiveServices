```python
key = "ff86cfd9565d4d03b2d5c1bd6294096c"
endpoint = "https://synergycognitivetest.cognitiveservices.azure.com/"

import csv
from azure.ai.textanalytics import TextAnalyticsClient
from azure.core.credentials import AzureKeyCredential
from csv import reader

# Authenticate the client using your key and endpoint 
def authenticate_client():
    ta_credential = AzureKeyCredential(key)
    text_analytics_client = TextAnalyticsClient(
            endpoint=endpoint, 
            credential=ta_credential)
    return text_analytics_client

client = authenticate_client()

# Example method for detecting the language of text
def language_detection_example(client, text):
    try:
        documents = [text]
        response = client.detect_language(documents = documents)[0]
        return response.primary_language.name

    except Exception as err:
        return "Encountered exception. {}".format(err)
    
```

```python
import pandas as pd
import numpy as np
results  = pd.read_csv('D:\TFM\dataset_multilanguage_result.csv')
results.loc[results["LanguageDetected"] == "Portuguese", "LanguageDetected"] = "Portugese"
results.loc[results["LanguageDetected"] == "Chinese_Simplified", "LanguageDetected"] = "Chinese"
results.loc[results["LanguageDetected"] == "Chinese_Traditional", "LanguageDetected"] = "Chinese"
results.loc[results["LanguageDetected"] == "Pashto", "LanguageDetected"] = "Pushto"
results.loc[results["LanguageDetected"] == "Farsi", "LanguageDetected"] = "Persian"
results['Compare'] = np.select([results ['Language'] != results ['LanguageDetected']], ['Different'], 'Same ')
results = results.drop(['Text'], axis=1)
countT = results.groupby(['Language','Compare']).size().reset_index(name="count")
results = results[results.Compare == 'Different']
print(results)
results.head(15)
```

```python
#results = (results[results.Compare == 'Different']).drop_duplicates()

count = results.groupby(['Language']).size().reset_index(name="count")
count = count.sort_values('count',ascending=True)
countT['countT'] = countT['count'].groupby(countT['Language']).transform('sum')
countT = countT[countT.Compare == 'Different']
countT = countT.sort_values('count',ascending=True)

#Formatear el dataframe en formato html
from IPython.core.display import HTML
HTML(countT.to_html(classes='table table-stripped'))
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
import pandas as pd
import matplotlib.pyplot as plt
results2  = pd.read_csv('D:\TFM\dataset_multilanguage_result.csv')
results2.loc[results2["LanguageDetected"] == "Portuguese", "LanguageDetected"] = "Portugese"
results2.loc[results2["LanguageDetected"] == "Chinese_Simplified", "LanguageDetected"] = "Chinese"
results2.loc[results2["LanguageDetected"] == "Chinese_Traditional", "LanguageDetected"] = "Chinese"
results2.loc[results2["LanguageDetected"] == "Pashto", "LanguageDetected"] = "Pushto"
results2.loc[results2["LanguageDetected"] == "Farsi", "LanguageDetected"] = "Persian"

cmd = ConfusionMatrixDisplay.from_predictions(results2['Language'], results2['LanguageDetected'])
fig, ax = plt.subplots(figsize=(15,15))
cmd.plot(ax=ax)
cmd.ax_.set(
                title='Confusion Matrix - Azure Cognitive Services', 
                xlabel='Predicted Values', 
                ylabel='Actual Values')

plt.xticks(rotation=90)
```
